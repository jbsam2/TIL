## 불쌍한 달팽이

> 달팽이는 낮 시간 동안에 기둥을 올라간다. 하지만 밤에는 잠을 자면서 어느 정도의 거리만큼 미끄러진다. (낮 시간 동안 올라간 거리보다는 적게 미끄러진다.)
> 
> 달팽이가 기둥의 꼭대기에 도달하는 날까지 걸리는 시간을 반환하는 함수 `snail()`을 작성하시오.

> 함수의 인자는 다음과 같다.
1. 기둥의 높이(미터)
2. 낮 시간 동안 달팽이가 올라가는 거리(미터)
3. 달팽이가 야간에 잠을 자는 동안 미끄러지는 거리(미터)

---

예시)

```python
snail(100, 5, 2) #=> 33
```

````python
#1
def snail(height, day, night):
  if height <= day:
    return 1
  else:
    up = day - night
    days = (height - day) // up
    if (height - day) % up:
      return days + 2
    else:
      return days + 1

#2
def snail(height, day, night):
  days = 0
  cur_height = 0
  while 1:
    days += 1
    cur_height += day
    if cur_height >= height:
      return days
      break
    cur_height -= night
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하시오.
print(snail(100, 5, 2))
````

## 자릿수 더하기 (SWEA #2058)

> 자연수 number를 입력 받아, 각 자릿수의 합을 계산하여 출력하시오.

---

예시)

```python
sum_of_digit(1234) #=> 10
sum_of_digit(4321) #=> 10
```

````python
#1
def sum_of_digit(number):
  ret = 0
  while number > 0:
    ret += (number % 10)
    number //= 10
  return ret

#2
def sum_of_digit(number):
  if number < 10:
    return number
  else:
    total = number % 10
    number //= 10
    return total + sum_of_digit(number)

#3
def sum_of_digit(number):
  num_str = str(number)
  ret = 0
  for i in num_str:
    ret += int(i)
  return ret
````

````python
# 해당 코드를 통해 올바른 결과가 나오는지 확인하시오.
print(sum_of_digit(1234))
print(sum_of_digit(4321))
````
